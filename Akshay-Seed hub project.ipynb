```python


class Seed:
    def __init__(self):
        self.seed_id = []
        self.name = []
        self.variety = []
        self.hybrid = []
        self.stock = []
        self.price = []

    def accept(self):
        seed_id = input("Enter the seed ID: ")
        name = input("Enter the name: ")
        variety = input("Enter the variety: ")
        hybrid = input("Is it a hybrid? (yes/no): ").lower() == 'yes'
        stock = int(input("Enter the stock quantity: "))
        price = float(input("Enter the price of the seed: "))

        self.seed_id.append(seed_id)
        self.name.append(name)
        self.variety.append(variety)
        self.hybrid.append(hybrid)
        self.stock.append(stock)
        self.price.append(price)

    def display(self):
        if not self.name:
            print("No seeds in inventory./n")
        else:
            print("Seed Store Inventory:/n")
            for i in range(len(self.name)):
                hybrid_status = "Hybrid Seed" if self.hybrid[i] else "Non-hybrid Seed"
                print(f"{self.seed_id[i]}-{self.name[i]}-{self.variety[i]}({hybrid_status}) - Stock:{self.stock[i]} - Price: {self.price[i]:.2f}")

    def restock(self, seed_id, quantity):
        index = self.seed_id.index(seed_id)
        self.stock[index] += quantity
        print(f"{quantity} {self.name[index]} seeds have been added to the stock.")

    def sell(self, seed_id, quantity):
        index = self.seed_id.index(seed_id)
        if self.stock[index] >= quantity:
            total_price = self.price[index] * quantity
            self.stock[index] -= quantity
            print(f"{quantity} {self.name[index]} seeds have been sold for a total of ${total_price:.2f}.")
            return seed_id, quantity, total_price
        else:
            print(f"Insufficient stock. Only {self.stock[index]} {self.name[index]} seeds available.")
            return None

class SeedStore:
    def __init__(self):
        self.seeds = Seed()
        self.sales_report = []  

    def add_seed(self):
        self.seeds.accept()
        print(f"{self.seeds.name[-1]} seeds have been added to the store.\n")

    def display_seeds(self):
        self.seeds.display()

    def generate_sales_report(self, sale_details):
        if sale_details:
          
            sale_entry = {
                
                'seed_id': sale_details[0],
                'quantity': sale_details[1],
                'total_price': sale_details[2]
            }
            self.sales_report.append(sale_entry)
            print("Sales report generated successfully.")

            print("\nSales Report:")
            for entry in self.sales_report:
                print(f"Seed ID: {entry['seed_id']} - Quantity: {entry['quantity']} - Total Price: ${entry['total_price']:.2f}")

    def restock(self, seed_id, quantity):
        self.seeds.restock(seed_id, quantity)

    def sell(self, seed_id, quantity):
        sale_details = self.seeds.sell(seed_id, quantity)
        self.generate_sales_report(sale_details)  

    def display_sales_report(self):
        if not self.sales_report:
            print("No sales report available.")
        else:
            print("\nSales Report:")
            for entry in self.sales_report:
                print(f"Seed ID: {entry['seed_id']} - Quantity: {entry['quantity']} - Total Price: $ {entry ['total_price'] :.2f}")

    def display_stock_list(self):
        print("\nStock List:")
        self.seeds.display()


seed_store = SeedStore()

while True:
    print('            ')
    print("           SEED HUB Pvt Ltd.        ")
    print("Please select an option below:")
    print("1. Add Seed\n2. Display Seeds\n3. Restock\n4. Sell\n5. Sales Report \n6. Stock Report\n7. Exit")
    choice = int(input())

    if choice == 1:
        seed_store.add_seed()
    elif choice == 2:
        seed_store.display_seeds()
    elif choice == 3:
        print("Before Restocking:")
        seed_store.display_seeds()
        seed_id = input("Enter the seed ID to restock: ")
        quantity = int(input("Enter the quantity to restock: "))
        seed_store.restock(seed_id, quantity)
        print("After Restocking:")
        seed_store.display_seeds()
    elif choice == 4:
        print("Before Selling:")
        seed_store.display_seeds()
        seed_id = input("Enter the seed ID to sell: ")
        quantity = int(input("Enter the quantity to sell: "))
        seed_store.sell(seed_id, quantity)
        print("After Selling:")
        seed_store.display_seeds()
    elif choice == 5:
        seed_store.display_sales_report()
    elif choice == 6:
         seed_store.display_stock_list()
    elif choice == 7:
        break
       
```
