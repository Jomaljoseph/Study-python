```python
class bees:
    gst=10/100
    def __init__(self):
        self.price_table={'honey':250,
                       'processed honey':300,
                       'forest honey':450,
                       'dates':120,
                       'cherry':150}
        self.stock={'honey':10,
                   'processed honey':10,
                   'forest honey':10,
                   'dates':10,
                   'cherry':10}
        self.fp=0
        
    def display_price_table(self):
        print(self.price_table)
        print('             ')
        
    def invoice(self):
        final_price=0
        discount_price=0
        print('select customer type')
        print('1.premium \n 2.regular \n 3.normal')
        customer=int(input())
        if customer==1:
            discount=5/100
        elif customer==2:
            discount=3/100
        elif customer==3:
            discount=0
        else:
            print('invalid input')
        print('            ')
        while True:  
            item=input('enter item name (type no to exit): ')
            if item.lower()=='no':
                break
            if item.lower() in self.price_table.keys():
                qty=int(input('enter quantity: '))
                total_price=self.price_table[item.lower()]*qty
                if item.lower()=='dates' or item.lower()=='cherry':  #
                    tax=total_price*bees.gst
                    total_price+=tax
                    print('GST: ',tax)
                if self.stock[item]>=qty:
                    self.stock[item]-=qty
                    print('total is: ',total_price)
                    print('               ')
                    final_price+=total_price
                    discount_price=final_price*discount
                    final_price-=discount_price
                else:
                    print('do not have enough stock for: ',item)
                    print('               ')
        self.fp+=final_price
        print('you are eligebile for a discount of: ',discount_price)
        print('final price is: ',final_price)
        print('                ')
        
    def display_stock(self):
        print(self.stock)
        print('                 ')
        
    def add_new_item(self):
        new_item=input('enter new item name: ')
        new_price=input('enter price: ')
        self.price_table.update({new_item:new_price})
        self.stock.update({new_item:0})
        print(self.price_table)
        print('             ')
    
    def add_stock(self):
        item=input('enter the item you want to add: ')
        quantity=int(input('enter quantity: '))
        self.stock[item]=self.stock[item]+quantity
        print(self.stock)
        print('         ')
        
    def collection(self):
        print('total collection is: ',self.fp)
        print('              ')
        
b=bees()
```

```python
while True:
    print("please select an option below")
    print("1.Display price table\n2.Billing\n3.Display stock\n4.Add new item\n5.Add stock\n6.Total collection")
    choice=int(input())
    if choice==1:
        b.display_price_table()
    elif choice==2:
        b.invoice()
    elif choice==3:
        b.display_stock()
    elif choice==4:
        b.add_new_item()
    elif choice==5:
        b.add_stock()
    elif choice==6:
        b.collection()
    else:
        print('invalid input, select a option 1 to 6')
        print('               ')
    
    while True:
        a=input("do you want to continue: ")
        print('                 ')
        if a.lower()=='yes': 
            break
        elif a.lower()=='no':
            break
        else:
            print('invalid input, select yes/no')
            continue
    if a.lower()=='no':
        break
```
