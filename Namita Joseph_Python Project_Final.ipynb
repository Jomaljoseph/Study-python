```python
from datetime import datetime
class ParkingLot:
    def __init__(self):
        self.pl_dict={}
        self.total_amount=0
        print("**********>>>>>NJ Parking Lot<<<<<**********")
    def checkin_vehicle(self):
        print("****CHECKIN VEHICLE****")
        v_id=int(input("Enter Vehicle ID: "))
        v_reg=input("Enter Registration Number: ")
        intime=datetime.now()
        self.pl_dict[v_id]={}
        self.pl_dict[v_id]['v_reg']=v_reg
        self.pl_dict[v_id]['vdate']=intime.strftime("%Y/%m/%d")
        self.pl_dict[v_id]['v_intime']=intime.strftime("%H:%M:%S")
        self.pl_dict[v_id]['v_outtime']="-"
        self.pl_dict[v_id]['status']="IN"
        self.pl_dict[v_id]['Parking_Amount']=10
        print("VehicleID",v_id,self.pl_dict[v_id])
        print()
    def checkout_vehicle(self):
        min_amount=10
        print("****CHECKOUT VEHICLE****")
        v_id=int(input("Enter Vehicle ID to send out: "))
        if (v_id not in self.pl_dict.keys()) or (self.pl_dict[v_id]['status']=="OUT"):
            print("Invalid Input. Check Vehicle ID")
        else:
            outtime=datetime.now()
            self.pl_dict[v_id]['v_outtime']=outtime.strftime("%H:%M:%S")
            self.pl_dict[v_id]['status']="OUT"
            t1=datetime.strptime(self.pl_dict[v_id]['v_intime'],"%H:%M:%S")
            t2=datetime.strptime(self.pl_dict[v_id]['v_outtime'],"%H:%M:%S")
            p_time=t2-t1
            total_time=(p_time.total_seconds())//10
            self.pl_dict[v_id]['Parking_Amount']+=(total_time*min_amount)
            self.total_amount+=self.pl_dict[v_id]['Parking_Amount']
            print("VehicleID",v_id,self.pl_dict[v_id])
            print()
            print("Bill is ",self.pl_dict[v_id]['Parking_Amount']," Rupees.")
            print()
    def in_vehicles(self):
        print("****IN VEHICLES****")
        for i in self.pl_dict.keys():
            if self.pl_dict[i]['status']=="IN":
                print("VehicleID",i,self.pl_dict[i]['v_reg'],"Date",self.pl_dict[i]['vdate'],"Time",self.pl_dict[i]['v_intime'])
        print()
    def out_vehicles(self):
        print("****OUT VEHICLES****")
        for i in self.pl_dict.keys():
            if self.pl_dict[i]['status']=="OUT":
                print("VehicleID",i,self.pl_dict[i]['v_reg'],"Date",self.pl_dict[i]['vdate'],"Time",self.pl_dict[i]['v_outtime'])
        print()
    def check_vehicle_status(self):
        print("****VEHICLE STATUS****")
        v_id=int(input("Enter Vehicle ID to check: "))
        if v_id not in self.pl_dict.keys():
            print("Invalid Input. Check Vehicle ID")
        else:
            print("VehicleID",v_id,"Registration Number",self.pl_dict[v_id]['v_reg'],"Status",self.pl_dict[v_id]['status'])
            print()
    def total_collection(self):
        print("****TOTAL COLLECTION****")
        print("Total Collection = ",self.total_amount," Rupees.")
        print()
                
v=ParkingLot()
while True:
    print("Enter your Choice")
    print("1. Check-IN Vehicle \n2. Check-OUT Vehicle\n3. Parked-IN Vehicles \n4. OUT Vehicles\n5. Check Vehicle Status\n6. Total Collection")
    choice=int(input())
    if choice==1:
        v.checkin_vehicle()
    elif choice==2:
        v.checkout_vehicle()
    elif choice==3:
        v.in_vehicles()
    elif choice==4:
        v.out_vehicles()
    elif choice==5:
        v.check_vehicle_status()
    elif choice==6:
        v.total_collection()
    else:
        print('!!Invalid input, select an Option from 1 to 6!!')
    
    while True:
        a=input("Do you want to continue (yes/no): ")
        print()
        if a.lower()=='yes': 
            break
        elif a.lower()=='no':
            break
        else:
            print('!!Invalid input, select yes/no!!')
            continue
    if a.lower()=='no':
        print("**********THANK YOU**********")
        break
```
